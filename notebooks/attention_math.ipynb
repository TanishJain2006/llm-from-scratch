# Attention Mechanism

## Based on the paper:
**Attention Is All You Need (Vaswani et al., 2017)**

### Objective
To understand the mathematical intuition behind the Attention Mechanism and implement its basic components using Python and NumPy.

## What is Attention?

Attention is a mechanism that allows a model to focus on the most relevant words in a sentence while processing a token.

Instead of treating every word equally, the model assigns higher importance to words that are more relevant to the current context.

## Why do we need Attention?

Before Transformers, models like RNNs and LSTMs processed words one by one.

As the sentence became longer, they struggled to remember important information.

Attention solves this problem by allowing every word to interact with every other word in the sentence.

## Query, Key and Value (Q, K, V)

Every token is converted into three vectors:

- Query (Q): What information am I looking for?
- Key (K): What information do I contain?
- Value (V): What information should I pass to the next layer?

```python
import numpy as np

Q = np.array([1, 0, 1])

K = np.array([1, 2, 0])

V = np.array([5, 3, 8])

print("Query :", Q)
print("Key :", K)
print("Value :", V)
```

## Step 1: Dot Product

The similarity between Query and Key is calculated using the Dot Product.

```python
score = np.dot(Q, K)

print("Dot Product =", score)
```

## Step 2: Scaling

The dot product is divided by the square root of the key dimension (√dk).

This prevents the values from becoming too large before applying Softmax.

```python
dk = len(K)

scaled_score = score / np.sqrt(dk)

print("Scaled Score =", scaled_score)
```

## Step 3: Softmax

Softmax converts attention scores into probabilities.

The probabilities always add up to 1.

```python
scores = np.array([1.2, 0.8, 2.1])

softmax = np.exp(scores) / np.sum(np.exp(scores))

print("Softmax Output:")
print(softmax)
```

## Final Attention Formula

The Transformer calculates attention using the following equation:

Attention(Q, K, V) = Softmax((QKᵀ / √dk)) V

Where:

- Q = Query
- K = Key
- V = Value
- dk = Dimension of Key

```python
weights = softmax

print("Attention Weights:")
print(weights)
```

## Observation

1. Every token generates Query, Key and Value vectors.
2. Query and Key are compared using the Dot Product.
3. The score is scaled using √dk.
4. Softmax converts the score into probabilities.
5. The probabilities are used to weight the Value vectors.
6. The final weighted values become the output of the Attention layer.
