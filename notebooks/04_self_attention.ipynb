# Self Attention Implementation

Based on the paper:
Attention Is All You Need (2017)

Objective:
Implement the Self Attention mechanism step by step using NumPy.

```python
import numpy as np
```

```python
import numpy as np
embeddings = np.array([
    [0.7, 0.8, 0.9],
    [0.4, 0.5, 0.6],
    [0.1, 0.2, 0.3]
])

print("Embedding Matrix:")
print(embeddings)
```

## Step 1: Create Query, Key and Value

In a Transformer, each embedding is converted into three vectors:

- Query (Q): What information is this token looking for?
- Key (K): What information does this token contain?
- Value (V): What information should be passed to the next layer?

For this simplified implementation, we will use the embedding matrix itself as Q, K and V.

```python
Q = embeddings

K = embeddings

V = embeddings

print("Query Matrix:")
print(Q)

print("\nKey Matrix:")
print(K)

print("\nValue Matrix:")
print(V)
```

## Step 2: Calculate Attention Scores

The first step in Self-Attention is to calculate how similar each Query is to every Key.

This is done using Matrix Multiplication:

Attention Scores = Q × Kᵀ

A higher score means the Query and Key are more similar.

```python
scores = np.matmul(Q, K.T)

print("Attention Scores:")
print(scores)
```

## Step 3: Scale the Attention Scores

The dot product values can become very large as the vector dimension increases.

To prevent this, the Transformer divides the attention scores by the square root of the key dimension (√dk).

This helps keep the values stable before applying the Softmax function.

```python
dk = K.shape[1]

scaled_scores = scores / np.sqrt(dk)

print("Scaled Attention Scores:")
print(scaled_scores)
```

## Step 4: Apply the Softmax Function

The scaled attention scores are converted into probabilities using the Softmax function.

Softmax ensures that:
- All values lie between 0 and 1.
- The sum of probabilities in each row is equal to 1.

These probabilities are called **Attention Weights**.

```python
def softmax(x):
    exp = np.exp(x - np.max(x, axis=-1, keepdims=True))
    return exp / np.sum(exp, axis=-1, keepdims=True)

attention_weights = softmax(scaled_scores)

print("Attention Weights:")
print(attention_weights)
```

## Step 5: Calculate the Final Attention Output

The attention weights are multiplied with the Value matrix.

This produces a new representation for each token by combining information from all the other tokens based on their importance.

```python
output = np.matmul(attention_weights, V)

print("Final Attention Output:")
print(output)
```

# Observation

The Self-Attention mechanism works as follows:

1. Convert input tokens into embedding vectors.
2. Create Query (Q), Key (K), and Value (V) matrices.
3. Compute attention scores using Q × Kᵀ.
4. Scale the scores by √dk.
5. Apply the Softmax function to obtain attention weights.
6. Multiply the attention weights with the Value matrix.
7. Generate the final attention output.

This is the core computation introduced in the paper **Attention Is All You Need (2017)**.
