# Embeddings

## Objective

To understand how token IDs are converted into embedding vectors.

```python
sentence = "I love AI"

print(sentence)
```

## Step 1

Tokenization

```python
tokens = sentence.split()

print(tokens)
```

## Step 2

Vocabulary

```python
vocabulary = {
    "I":0,
    "love":1,
    "AI":2
}

print(vocabulary)
```

## Step 3

Token IDs

```python
token_ids = [vocabulary[token] for token in tokens]

print(token_ids)
```

## Step 4

Embedding Matrix

```python
embedding_matrix = {
    0:[0.7,0.8,0.9],
    1:[0.4,0.5,0.6],
    2:[0.1,0.2,0.3]
}

print(embedding_matrix)
```

## Step 5

Embedding Vectors

```python
embedding_matrix = {
    0:[0.7,0.8,0.9],
    1:[0.4,0.5,0.6],
    2:[0.1,0.2,0.3]
}

embeddings = [embedding_matrix[id] for id in token_ids]

for token, vector in zip(tokens, embeddings):
    print(token, ":", vector)
```

## Observation

Raw text cannot be directly processed by an LLM.

The text is first tokenized.

Each token is converted into an ID.

Each ID is mapped to an embedding vector.

These vectors are then passed to the transformer.
