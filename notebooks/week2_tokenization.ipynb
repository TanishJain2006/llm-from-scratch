```python
text = "I love artificial intelligence"

tokens = text.split()

print(tokens)
```

```python
text = "Hello"

tokens = list(text)

print(tokens)
```

```python
text = "I love AI and I love coding"

tokens = text.split()

vocab = sorted(set(tokens))

print(vocab)
```

```python
vocab = ['AI', 'I', 'and', 'coding', 'love']

word_to_id = {}

for i, word in enumerate(vocab):
    word_to_id[word] = i

print(word_to_id)
```

```python
text = "I love AI"

tokens = text.split()

encoded = [word_to_id[word] for word in tokens]

print(encoded)
```

## Observations

1. Text cannot be directly processed by an LLM.
2. Text is first broken into tokens.
3. A vocabulary is created from unique tokens.
4. Each token is mapped to a numerical ID.
5. The model processes numerical IDs instead of raw text.
